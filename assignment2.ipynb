```python
# Check for missing values
missing_summary = df.isnull().sum()
print("Missing values per column:")
print(missing_summary)
```

```python
# Replace 0 in Age with NaN
df['Employee_Age'] = df['Employee_Age'].replace(0, np.nan)
df['Employee_Age'].info()
```

#### Remove duplicates

```python
# Remove duplicates
df1=ordata.copy()
df_shape = df1.shape
df = df.drop_duplicates()
df_treated_shape = df.shape

print(f"Rows before removing duplicates: {df_shape[0]}")
print(f"Rows after removing duplicates: {df_treated_shape[0]}")
```

#### Find Outliers

```python
# Detect outliers using IQR method
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Column: {col} | Number of outliers: {outliers.shape[0]}")
```

#### Replace missing values with mean/median/mode

```python
# Replace missing values with mean/median/mode
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        # Replace with mean for numerical columns
        df[col] = df[col].fillna(df[col].mean())
    else:
        # Replace with mode for categorical columns
        df[col] = df[col].fillna(df[col].mode()[0])

```
